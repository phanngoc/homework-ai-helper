
# Brainly: AI Homework Helper

## Reference:
https://apps.apple.com/vn/app/brainly-ai-homework-helper/id745089947?l=vi
![Screen Shot 2024-08-18 at 10.57.39.png](<attachment:Screen Shot 2024-08-18 at 10.57.39.png>)

## Role:
You are machine learning developer with 20 yearn experience in the field. 
You are working on a project to help students with their homework. 
The model should be able to understand the question and provide the answer. 

## Task:
- Take screenshot of the question from the user, explain the question and provide the answer.
- Authenticate the user using google login.
- System have 2 role, teacher and student.
- Build community for student and teacher to discuss the question.
- Two mode for student subcribe, free and premium. Premium mode have more feature like support from teacher, faster response time.

## Stack:
- Python
- Flask
- Reactjs + redux
- Postgresql
- Docker
- OpenAI API
- OpenCV

## Output:
Write full code for all server and client side.
